```python
import folium
import requests

# Replace 'YOUR_API_KEY' with your actual Geoapify API key
api_key = 'eb090fa5ec57436fab6b694e5bc43ee3'



# Define the points
origin = (37.7749, -122.4194)
waypoint1 = (37.7833, -122.4167)
waypoint2 = (37.7946, -122.3994)
destination = (37.7896, -122.3901)

# Send requests to the Geoapify Routing API to get the routes
waypoints = [origin, waypoint1, waypoint2, destination]
route_features = []

for i in range(len(waypoints) - 1):
    url = f'https://api.geoapify.com/v1/routing?waypoints=lonlat:{waypoints[i][1]},{waypoints[i][0]}|lonlat:{waypoints[i+1][1]},{waypoints[i+1][0]}&apiKey={api_key}&mode=drive&format=geojson'
    response = requests.get(url)
    data = response.json()
    route_features.append(data['features'][0])

# Create a map centered at the origin
m = folium.Map(location=origin, zoom_start=12)

# Add markers for the origin, waypoints, and destination
folium.Marker(origin, tooltip='Origin').add_to(m)
folium.Marker(waypoint1, tooltip='Waypoint 1').add_to(m)
folium.Marker(waypoint2, tooltip='Waypoint 2').add_to(m)
folium.Marker(destination, tooltip='Destination').add_to(m)

# Create a GeoJSON layer for the routes
route_layer = folium.GeoJson(data={
    "type": "FeatureCollection",
    "features": route_features
}, name='Routes')

# Add the route layer to the map
route_layer.add_to(m)

# Display the map
m
```
