```python
import folium
import requests

# Your API key
myAPIKey = "eb090fa5ec57436fab6b694e5bc43ee3"

# Waypoints
fromWaypoint = [38.937165, -77.045590]  # latitude, longitude
toWaypoint = [38.881152, -76.990693]    # latitude, longitude

# Call Routing API
url = f"https://api.geoapify.com/v1/routing?waypoints={fromWaypoint[0]},{fromWaypoint[1]}|{toWaypoint[0]},{toWaypoint[1]}&mode=drive&details=instruction_details&apiKey={myAPIKey}"
response = requests.get(url)
routeResult = response.json()

# Print the route result for inspection
print(routeResult)

# Get the route geometry
route_geometry = routeResult.get('geometry', {}).get('coordinates', [])

if not route_geometry:
    raise ValueError("Route geometry is empty")

# Create a map centered at the first waypoint
map_center = [fromWaypoint[0], fromWaypoint[1]]
map = folium.Map(location=map_center, zoom_start=12)

# Add markers for the waypoints
folium.Marker(fromWaypoint, popup='From Waypoint').add_to(map)
folium.Marker(toWaypoint, popup='To Waypoint').add_to(map)

# Add the route to the map as a Polyline
folium.PolyLine(locations=route_geometry, color='blue', weight=5, opacity=0.7).add_to(map)

# Display the map
map.save('route_map.html')
```

```python
import folium
import requests

# API Key
my_api_key = "eb090fa5ec57436fab6b694e5bc43ee3"

# Create map centered at a specific location
my_map = folium.Map(location=[44.53361448499783, -72.79419772520356], zoom_start=14)

# Calculate and display routing
from_waypoint = [44.53000255267429, -72.78056761690857]  # [latitude, longitude]
to_waypoint = [44.536552001130076, -72.80797096598127]   # [latitude, longitude]

# Add markers for from and to waypoints
folium.Marker(location=from_waypoint, popup="Start").add_to(my_map)
folium.Marker(location=to_waypoint, popup="End").add_to(my_map)

# Fetch route data from Geoapify API
url = f"https://api.geoapify.com/v1/routing?waypoints=lonlat:{from_waypoint[1]},{from_waypoint[0]}|lonlat:{to_waypoint[1]},{to_waypoint[0]}&mode=hike&details=route_details,elevation&apiKey={my_api_key}"
response = requests.get(url)
route_data = response.json()

# Extract route coordinates
route_coordinates = []
for leg in route_data['features'][0]['properties']['legs']:
    leg_coords = route_data['features'][0]['geometry']['coordinates'][leg['from_index']:leg['to_index']+1]
    route_coordinates.extend(leg_coords)

# Add route polyline to the map
folium.PolyLine(locations=route_coordinates, color="blue", weight=2.5, opacity=1).add_to(my_map)

# Display the map
my_map
```
